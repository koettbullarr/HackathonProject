```python
def data_cleaner(dataset):
    
    df = dataset.copy()
    df['Detected Bots or Trolls'] = df['Detected Bots or Trolls'].apply(lambda x: True if x == "Possible bot activity detected" else False)
    
    df['Fact-Check Source'] = df['Fact-Check Source'].apply(lambda x: "unknown" if pd.isna(x) else x)
    
    return df
    
    
```

```python
df = data_cleaner(df)
df.info()
```
